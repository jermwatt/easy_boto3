```python
import subprocess

def pull_repo(repo_url):
    repo_name = repo_url.split('/')[-1].split('.')[0]
    
    # SSH alias details
    alias_name = 'shift'

    # Construct the SSH command
    ssh_command = f'ssh {alias_name} '
    
    # Execute commands on the remote server
    commands = [
        'cd /home/ubuntu',
        f'git clone {repo_url}',
        f'cd /home/ubuntu/{repo_name} && git checkout double_stochastic_test',
        f'cd /home/ubuntu/{repo_name} && sudo pip3 install -r requirements.txt',
        f'sudo chmod -R 777 /home/ubuntu/{repo_name}'
    ]

    # Execute commands sequentially
    for command in commands:
        full_command = ssh_command + '"' + command + '"'
        print(full_command)
        subprocess.run(full_command, shell=True, check=True)


    # Close the SSH connection
    ssh_client.close()
```
